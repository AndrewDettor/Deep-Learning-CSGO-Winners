### Drop

```python
dmg.columns
```

### Aggregation (per file, per round, per side):
- number of instances of damage
- total damage
- number of headshots
- number of instances of damage per weapon type
- if bomb was planted

```python
dmg["headshot"] = dmg["hitbox"].apply(lambda x: x == "Head")
dmg = dmg.drop(["hitbox"], axis=1)
dmg.sample(3)
```

```python
dmg["damage"] = dmg["hp_dmg"] + dmg["arm_dmg"]
dmg = dmg.drop(["hp_dmg", "arm_dmg"], axis=1)
dmg.sample(3)
```

```python
for wp in dmg["wp_type"].unique():
    dmg[f"wp_type_{wp}"] = dmg["wp_type"] == wp

dmg = dmg.drop(["wp_type"], axis=1)
dmg.sample(3)
```

```python
dmg = dmg[dmg["att_side"] != "None"]
dmg.sample(3)
```

```python
dmg["num_dmg_instances"] = 1
```

```python
agg_dict = {"is_bomb_planted": "any", 
            "headshot": "sum",
            "damage": "sum",
            "num_dmg_instances": "sum",
            "wp_type_Unkown": "sum",
            "wp_type_Pistol": "sum",
            "wp_type_Grenade": "sum",
            "wp_type_Rifle": "sum",
            "wp_type_Heavy": "sum",
            "wp_type_SMG": "sum",
            "wp_type_Sniper": "sum",
            "wp_type_Equipment": "sum",}
```

```python
dmg_g = dmg.groupby(["file", "round", "att_side"]).agg(agg_dict)
dmg_g
```

```python
dmg_g = dmg_g.rename(columns={"headshot": "headshot_hits"})
```

```python
wp_cols = [col for col in dmg_g.columns if "wp_type" in col]
wp_cols
```

```python
dmg_g = dmg_g.rename(columns={col:col+"_damage" for col in wp_cols})
```

### Aggregation (per file, per round, per side):
- number of grenades thrown
- number of each grenade thrown
- if bomb was planted

```python
grenades.sample(3)
```

```python
for nade in grenades["nade"].unique():
    grenades[f"{nade}_thrown"] = grenades["nade"] == nade

grenades = grenades.drop(["nade"], axis=1)
grenades.sample(3)
```

```python
grenades = grenades.rename(columns={"Incendiary_thrown": "Incendiary_hits", "Molotov_thrown": "Molotov_hits", "HE_thrown": "HE_hits"})
grenades.sample(3)
```

```python
agg_dict = {"is_bomb_planted": "any",
            "Smoke_thrown": "sum",
            "HE_hits": "sum",
            "Molotov_hits": "sum",
            "Flash_thrown": "sum",
            "Incendiary_hits": "sum",
            "Decoy_thrown": "sum",
            }
```

```python
grenades_g = grenades.groupby(["file", "round", "att_side"]).agg(agg_dict)
grenades_g
```

```python
grenades_g = grenades_g.rename(columns={"Smoke_thrown": "Smokes_thrown", "Flash_thrown": "Flashes_thrown", "Decoy_thrown": "Decoys_thrown"})
```

```python
import math
import pandas as pd
import numpy as np
```

***

```python
df = pd.read_csv("csgo_cleaned.csv")
df.info()
```

```python
cast_to_float = df.columns[4:]
df[cast_to_float] = df[cast_to_float].astype("float")
```

```python
files = df["file"].unique()
files[0]
```

```python
df[df["file"] == files[0]]
```

```python
# limit to no overtime rounds
df = df.loc[df["round"] <= 30]
```

```python
# drop map
# df = df.drop("map", axis=1)
```

```python
# round-wise average
mean_per_round = df.groupby("round").mean()
mean_per_round
```

```python
df.shape
```

```python
df.columns
```

****

```python
# find which rounds are missing per game
missing_df = pd.DataFrame(columns=df.columns)
for file in files:
    df_f = df.loc[df["file"] == file]
    df_r_np = df_f["round"].to_numpy().flatten()
    last_r = df_r_np[-1]
    should_be = np.arange(1, last_r+1)
    
    # what rounds are missing in this match?
    missing = list(set(should_be) - set(df_r_np))
    
    map_ = df_f["map"].to_numpy().flatten()[0]
    assert isinstance(map_, str), "map bad"
    
    # create new rows to append to big df
    for round_ in missing:
        avg = list(mean_per_round.loc[round_, :].to_numpy().flatten())
        new_row = [file, map_, round_, np.nan] + avg
        missing_df.loc[missing_df.shape[0], :] = new_row

missing_df["was_missing"] = 1

missing_df
```

```python
df["was_missing"] = 0
```

```python
# append missing rounds to df and re-sort
df_imputed = pd.concat([df, missing_df], axis=0)
df_imputed = df_imputed.sort_values(by=["file", "round"])
df_imputed = df_imputed.reset_index(drop=True)

df_imputed[df_imputed["file"] == files[0]]
```

```python
df_imputed.loc[df_imputed["file"] == files[0], "winner_side"].shift(periods=-1)
```

```python
# shift winner side back 1 (y)
for file in files:
    df_imputed.loc[df_imputed["file"] == file, "next_round_winner"] = df_imputed.loc[df_imputed["file"] == file, "winner_side"].shift(periods=-1, fill_value=np.nan)
df_imputed[df_imputed["file"] == files[0]]
```

```python
df_imputed.loc[df_imputed["file"] == files[0], ["file", "round", "winner_side", "next_round_winner", "was_missing"]]
```

```python
df_imputed.to_csv("csgo_imputed_shifted.csv", index=False)
```
